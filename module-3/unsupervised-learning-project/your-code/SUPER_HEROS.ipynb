Comic characters data

The raw data behind the story Comic Books Are Still Made By Men, For Men And About Men.

The data comes from Marvel Wikia and DC Wikia. Characters were scraped on August 24. Appearance counts were scraped on September 2. The month and year of the first issue each character appeared in was pulled on October 6.

The data is split into two files, for DC and Marvel, respectively: dc-wikia-data.csv and marvel-wikia-data.csv. Each file has the following variables:
Variable 	Definition
page_id 	The unique identifier for that characters page within the wikia
name 	The name of the character
urlslug 	The unique url within the wikia that takes you to the character
ID 	The identity status of the character (Secret Identity, Public identity, [on marvel only: No Dual Identity])
ALIGN 	If the character is Good, Bad or Neutral
EYE 	Eye color of the character
HAIR 	Hair color of the character
SEX 	Sex of the character (e.g. Male, Female, etc.)
GSM 	If the character is a gender or sexual minority (e.g. Homosexual characters, bisexual characters)
ALIVE 	If the character is alive or deceased
APPEARANCES 	The number of appareances of the character in comic books (as of Sep. 2, 2014. Number will become increasingly out of date as time goes on.)
FIRST APPEARANCE 	The month and year of the character's first appearance in a comic book, if available
YEAR 	The year of the character's first appearance in a comic book, if available

```python
heroes_information = pd.DataFrame(pd.read_csv('heroes_information.csv'))
heroes_information.drop(columns='Unnamed: 0', inplace=True)
heroes_information.rename(columns={"name": "hero_names"}, inplace=True)
```

```python
powers_list=[col for col in super_hero_powers]
powers_list.remove('hero_names')
powers_list
```

Pour le projet de clustering, je fusionne les deux df par le nom du heros.

```python
super_hero= pd.merge(heroes_information,
                     super_hero_powers,
                     on='hero_names')
super_hero.head()
```

```python
super_hero.set_index('hero_names', inplace=True)
```

Pour les colonnes de pouvoir, le genre, l'"Alignment", je peux facilement transformer les colonnes de type Object et Booléen en colonnes numériques (ordinal)

### les colonnes de type booléen

Les colonnes de type booléen sont celle qui identifient les pouvoirs des super-héros.
Je choisis de mettre 1 si le super-héros a un pouvoir et 0 s'il ne l'a pas.

```python
for power in powers_list:
    super_hero[power] = np.where(super_hero[power] == True, 1,0)
```

### La colonne Gender
Male=1, Female=2, '-'=0

```python
super_hero['Gender'] = np.where(super_hero['Gender'] == 'Male', 1,
                         (np.where(super_hero['Gender']=='Female', 2,0)))
```

### La colonne Alignment
'good'=1, 'bad'=2, '-'=0 ,'neutral'=3

```python
super_hero['Alignment'] = np.where(super_hero['Alignment'] == 'good', 1,
                         (np.where(super_hero['Alignment']=='bad', 2,(np.where(super_hero['Alignment']=='neutral', 3,0)))))
```

```python
for col in super_hero:
    print(col,':',super_hero[col].unique())
```
