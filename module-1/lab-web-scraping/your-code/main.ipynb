````python
#your code
def get_nb_tweets(url):
    try:
        twitter = requests.get(url).text
    except Exception :
        print('count not found')
        return

    hack = BeautifulSoup(twitter)
    nb_tweet=hack.find_all('a', {'data-nav':'tweets'})
    for x in nb_tweet:
        print(x.text)
get_nb_tweets(url)
````
