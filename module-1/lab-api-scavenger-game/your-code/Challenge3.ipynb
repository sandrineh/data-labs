```python
# your code here
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
#import re
#import base64
```

```python
h = open('lab_scavenger.txt').read()
headers=h

##### https://github.com/ironhack-datalabs/scavenger
#### GET /repos/:owner/:repo/git/trees/:tree_sha

url_tree='https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/0d82bf748275dbc1239a11e9157b198d88da7383'

trees = requests.get(url_tree,headers)
trees.json()
```

```python
df_trees = pd.DataFrame(trees.json())
```

```python
df_trees.head(10)
```

```python
df_trees.columns
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data

nested_columns = ['tree']

flat_tree = flatten(df_trees, nested_columns)
flat_tree.head(10)
```

```python
#GET /repos/:owner/:repo/contents/:path
#https://api.github.com/repos/ironhack-labs/data-labs/contents/
```

```python
df_content = pd.DataFrame()

for path in flat_tree['tree_path']:
    url_path=f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{path}'
    content = requests.get(url_path,headers)
    j_content = content.json()
    df = json_normalize(j_content)
    df_content = df_content.append(df,sort=True)
df_content.head()
```

```python
df_content['name'].value_counts()
```

```python
result_list =df_content.loc[(df_content['name'].str.contains('scavengerhunt')==True)]
result_list
```

```python
joke = result_list.sort_values(by='name', ascending=True)
joke
```

```python
list_joke=[]
for word in joke['download_url']:
    joke_word=requests.get(word).content.decode("utf-8")
    list_joke.append(joke_word)
list_joke

joke_final =' '.join(l[:-1] for l in list_joke)

joke_final
```
