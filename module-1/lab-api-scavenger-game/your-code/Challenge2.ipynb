```python
# your code here
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
h = open('lab_scavenger.txt').read()
headers=h
```

```python
#commits past week
"""
info obtenue avec $Curl pour connaître le nombre total de pages pour les commits

Link: <https://api.github.com/repositories/147683540/commits?page=2>; rel="next", 
         <https://api.github.com/repositories/147683540/commits?page=15>; rel="last"
"""

df_commit = pd.DataFrame()

total_pages=15
for i in range(total_pages):
    url=f'https://api.github.com/repos/ironhack-labs/data-labs/commits?page={i}&since=2019-07-01T00:00:00Z&until=2019-07-14T23:59:59Z'
    commits = requests.get(url,headers)
    j_commits = commits.json()
    df = json_normalize(j_commits)
    df_commit = df_commit.append(df, sort=True)

df_commit
```
